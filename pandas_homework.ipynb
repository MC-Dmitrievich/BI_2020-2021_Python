```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns  
```

```python
path = "https://raw.githubusercontent.com/Serfentum/bf_course/master/14.pandas/train.csv"
df = pd.read_csv(path)
df
```

```python
# Problem 1
df2 = df[['pos', 'A', 'C', 'T', 'G']]
plt.rcParams['figure.figsize'] = [20, 8] 
df2.plot.bar(x='pos', stacked=True)
```

```python
# Problem 2 
matches_mean = df.matches.mean() 
df3 = df.loc[df.matches > matches_mean]
df4 = df3[['pos', 'reads_all', 'mismatches', 'deletions', 'insertions']]
df4.to_csv('train_part.csv')
df4
```
